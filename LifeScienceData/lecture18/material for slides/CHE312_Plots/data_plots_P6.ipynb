```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting defaults for a presentation
# explore on the first graph how the plot changes when you switch the context to e.g. "paper" or "poster"
sns.set_theme(style='ticks', context='talk') #talk is default
plt.rcParams['figure.dpi'] = 120
```

```python
# load both data sets (TLC and HPLC)
gallium_68_SAR_TLC_df = pd.read_csv("richtige gooner Run 30.csv")

gallium_68_SAR_TLC_df.head()
```

```python
gallium_68_SAR_HPLC_df = pd.read_csv("DOMY_JN-1-68Ga 2 - CH1.txt", sep="\s+")  # automatic whitespace separator

gallium_68_SAR_HPLC_df.head()
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=gallium_68_SAR_TLC_df,
    x='Distance',
    y='Radioactivity',     
    linewidth=1
)

sns.set_palette("colorblind")

xmin = gallium_68_SAR_TLC_df['Distance'].min()
xmax = gallium_68_SAR_TLC_df['Distance'].max()
plt.xlim(xmin, xmax)

ymin = gallium_68_SAR_TLC_df['Radioactivity'].min()
ymax = gallium_68_SAR_TLC_df['Radioactivity'].max()
plt.ylim(ymin, ymax)

sns.despine()


plt.xlabel(r'$\bf{Retention~[mm]}$')
plt.ylabel(r'$\bf{Counts}$')
plt.tight_layout()
plt.savefig("Radio_TLC_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Scatter + mean trend with uncertainty - 2 plots in one canvas
plt.figure(figsize=(10,6))

sns.lineplot(
    data=gallium_68_SAR_HPLC_df,
    x='Time',
    y='Intensity',     
    linewidth=1
)

sns.set_palette("colorblind")

xmin = gallium_68_SAR_HPLC_df['Time'].min()
xmax = gallium_68_SAR_HPLC_df['Time'].max()
plt.xlim(xmin, xmax)

ymin = gallium_68_SAR_HPLC_df['Intensity'].min()
ymax = gallium_68_SAR_HPLC_df['Intensity'].max()
plt.ylim(ymin, ymax)

sns.despine()

plt.tick_params(axis='both', which='both')

plt.xlabel(r'$\bf{Retention~time~[min]}$')
plt.ylabel(r'$\bf{Intensity}$')
plt.tight_layout()
plt.savefig("Radio_HPLC_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```
